```python
def chatbot():
    while True:
        text = input("You: ").strip().lower()
        if text in ("quit", "exit"):
            print("Bot:GoodBye!")
            break
        elif "hello" in text:
            print("Bot :Hi! how can i help you ?")
        else:
            print("Bot: i don't understand")
if __name__ == "__main__":
    chatbot()
```
